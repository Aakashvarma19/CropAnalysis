**1ST FUNCTIONALITY - ARM ALGORITHM**

**IMPORTING LIBRARIES**:


```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

**IMPORTING** **DATASET** :

```python
a=pd.read_csv("/kerala.csv")
a=a.iloc[:,1:]
a.head()
a=a[a['Crop_Year']<2003]
a=a[a['Production']>1000]
a.head()

```

**DATA PREPROCESSING**:

```python
table1 = (a.groupby(['District_Name', 'Crop'])['Production']
          .sum().unstack().reset_index().fillna(0)
          .set_index('District_Name'))
table1
```

**TRANSACTION TABLE GENERATION:**

```python
def trnsction(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
table2=table1.applymap(trnsction)
table2
```

**FORMING FREQUENT ITEMSETS:**

```python
aprori = apriori(table2, min_support = 0.5, use_colnames = True)
aprori
```

**ASSOCIATION RULE:**

```python
arm = association_rules(aprori, metric ="lift", min_threshold = 1)
arm = arm.sort_values(['confidence', 'lift'], ascending =[False, False])
arm
```

**2nd FUNCTIONALITY - DECISION TREE**

**IMPORTING LIBRARIES**


```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn import tree
```

**DATA PREPROCESSING**

```python
b=pd.read_csv("/content/Fertilizer Prediction.csv")
p=b.iloc[:,:-1]
q=b.iloc[:,-1]
p1=p
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3,4])], remainder='passthrough')
p = np.array(ct.fit_transform(p))
```

```python
train_set1,test_set1,train_set2,test_set2=train_test_split(p,q,test_size=0.3,random_state=2)
```

**DECISION TREE GENERATION**

**ENTROPY METHOD**

```python
troi=DecisionTreeClassifier(criterion='entropy',random_state=2)
troi.fit(train_set1,train_set2)
```

**ACCURACY PREDICTION**

```python
prediction=troi.predict(test_set1)
accuracy_score(test_set2,prediction)
```

**GINI INDEX METHOD**

```python
troi1=DecisionTreeClassifier(criterion='gini',random_state=2)
troi1.fit(train_set1,train_set2)
```

```python
prediction=troi1.predict(test_set1)
accuracy_score(test_set2,prediction)
```
